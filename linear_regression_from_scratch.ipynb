```
from random import seed
from random import randrange
from csv import reader
from math import sqrt

# loading csv file
def load_csv(filename):
  dataset=list()
  with open(filename,'r') as file:
    csv_reader=reader(file)
    for row in csv_reader:
      if not row:
        continue
      else:
        dataset.append(row)
  return dataset

# Convert string column to float
def str_column_to_float(dataset,col):
  length=len(dataset)
  dataset=dataset[1:length]
  dataset_copy=list()
  for e in range(len(dataset)):
    if len(dataset[e])==col:
      temp=[]
      for k in range(col):
        temp.append(float(dataset[e][k]))
      dataset_copy.append(temp)
  return dataset_copy


# Split a dataset into a train and test set
def train_test_split(dataset,split):
  train=list()
  train_size=split*len(dataset)
  dataset_copy=list(dataset)
  while len(train)<train_size:
    index=randrange(len(dataset_copy))
    train.append(dataset_copy.pop(index))
  return train,dataset_copy


# Calculate root mean squared error
def rmse_metric(actual, predicted):
	sum_error = 0.0
	for i in range(len(actual)):
		prediction_error = predicted[i] - actual[i]
		sum_error += (prediction_error ** 2)
	mean_error = sum_error / float(len(actual))
	return sqrt(mean_error)



def evaluate_algorithm(dataset,test_set, algorithm, split, *args):
  train,test=train_test_split(dataset,split)
  test_set1=list()
  '''
  for row in test:
    row_copy=list(row)
    row_copy[-1]=None
    test_set1.append(row_copy)
    
  test_set2=list()
  for row in test_set:
    row_copy=list(row)
    row_copy[-1]=None
    test_set2.append(row_copy)
  '''
    
    
  predicted_train=algorithm(train,test,*args)
  print(predicted_train)
  
  predicted_test=algorithm(train,test_set,*args)
  print(predicted_test)
  
  actual_train=[row[-1] for row in test]
  actual_test=[row[-1] for row in test_set]
  
  rmse_train=rmse_metric(actual_train,predicted_train)
  rmse_test=rmse_metric(actual_test,predicted_test)
  
  return rmse_train,rmse_test



# calculate mean value of data
def mean(value):
  return sum(value) / float(len(value))


# claculate variance of data
def variance(value,mean):
  return sum([((x-mean)**2) for x in value])



# Calculate covariance between x(attribute) and y(label)
def covariance(x, mean_x, y, mean_y):
	covar = 0.0
	for i in range(len(x)):
		covar += (x[i] - mean_x) * (y[i] - mean_y)
	return covar


# Calculate coefficients
def coefficients(dataset):
	x = [row[0] for row in dataset]
	y = [row[1] for row in dataset]
	x_mean, y_mean = mean(x), mean(y)
	b1 = covariance(x, x_mean, y, y_mean) / variance(x, x_mean)
	b0 = y_mean - b1 * x_mean
	return [b0, b1]


# Simple linear regression algorithm
def simple_linear_regression(train, test):
  predictions=list()
  b0,b1 = coefficients(train)
  for row in test:
    yhat=b0+b1*row[0]
    predictions.append(yhat)
  return predictions

#preparing dataset
filename="/content/gdrive/My Drive/linear_reg_from_scratch/"
dataset = load_csv(filename+"train.csv")
test_set= load_csv(filename+"test.csv")

col=len(dataset[0])
dataset=str_column_to_float(dataset,col)
col1=len(test_set[0])
test_set=str_column_to_float(test_set,col)

# evaluation and prediction
split = 0.6
rmse_train,rmse_test = evaluate_algorithm(dataset,test_set, simple_linear_regression, split)
print('RMSE_train: %.3f' % (rmse_train))
print('RMSE_test: %.3f' % (rmse_test))
```
